## roda o pipeline

```python
res = analyze_continuous_features(
    df= df,
    target_col= "sales_qty",
    reports_dir= reports_dir,
    project_dir= ".",
    min_nonnull_frac= 0.6,
    n_neighbors_mi=  3,
    random_state=  42,
    filename_prefix="feature_2_"
)
```

```python
df_filtred = res["rank_df"]
df_filtred = df_filtred[(df_filtred['Classificação Pearson'] != "Baixo") | (df_filtred['Classificação Mi'] != "Baixo") ]

individually_relevant_features = df_filtred["feature"]
```
