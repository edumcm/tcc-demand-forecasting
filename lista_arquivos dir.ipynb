```python
import os

def print_directory_structure(root_dir, prefix=""):
    """
    Percorre recursivamente todas as pastas e arquivos a partir de root_dir,
    imprimindo uma estrutura hierárquica no formato tipo árvore.
    """
    # Lista todos os itens no diretório atual e ordena (pastas primeiro)
    items = sorted(os.listdir(root_dir))
    entries = [item for item in items if not item.startswith('.')]  # ignora arquivos ocultos

    for index, entry in enumerate(entries):
        path = os.path.join(root_dir, entry)
        connector = "└─ " if index == len(entries) - 1 else "├─ "

        print(prefix + connector + entry)

        # Se for diretório, faz a chamada recursiva
        if os.path.isdir(path):
            extension = "   " if index == len(entries) - 1 else "│  "
            print_directory_structure(path, prefix + extension)

# --- Exemplo de uso ---
# Caminho raiz (pode ser "." se o notebook está na mesma pasta do projeto)
root_path = "/content/drive/MyDrive/tcc-modelo/tcc-demand-forecasting"
print(f"Estrutura de diretórios a partir de: {os.path.abspath(root_path)}\n")
print_directory_structure(root_path)
```
